
We can see that our data has no missing or strange values in it this means that we can go straight to creating our model

but first let's change names of columns so we can have better understanding of your data

```python
df = raw_data
```

```python
df.head()
```

```python
df.columns = ['age', 'sex', 'chest_pain_type', 'resting_blood_pressure', 'cholesterol', 'fasting_blood_sugar', 'rest_ecg', 'max_heart_rate_achieved',
       'exercise_induced_angina', 'st_depression', 'st_slope', 'num_major_vessels', 'thalassemia', 'target']
df.columns
```

```python
df['sex'] = df['sex'].map({1:'male',0:'female'})

df['chest_pain_type'] = df['chest_pain_type'].map({0:'typical angina',1:'atypical angina',
                                                   2:'non-anginal pain',3:'asymptomatic'})

df['fasting_blood_sugar'] = df['fasting_blood_sugar'].map({0:'lower than 120mg/ml',
                                                          1:'greater than 120mg/ml'})

df['rest_ecg'] = df['rest_ecg'].map({0:'normal',1:'ST-T wave abnormality',2:'left ventricular hypertrophy'})

df['exercise_induced_angina'] = df['exercise_induced_angina'].map({0:'no',1:'yes'})

df['st_slope'] = df['st_slope'].map({0:'upsloping',1: 'flat',2:'downsloping'})

df[df['thalassemia']==0] = 1
df['thalassemia'] = df['thalassemia'].map({1: 'normal',2: 'fixed defect',3: 'reversable defect'})
```

```python
df.head()
```

```python
df.info()
```

## Let's create our model

```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
X = pd.get_dummies(X,drop_first=True)
X
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
classifier = RandomForestClassifier(n_estimators = 500)
```

```python
classifier.fit(X_train,y_train)
```

```python
predictions = classifier.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
print('\n')
print(confusion_matrix(y_test,predictions))
```

## Let's see what we can do using neural networks

```python
 #X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
 #type(X_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(mode='min',patience=3,verbose=1)
```

```python
model = Sequential()
```

```python
model.add(Dense(24,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(24,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation = 'sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam') 
```

```python
y_train = y_train.values
```

```python
model.fit(X_train,y_train,batch_size=32,epochs = 100,verbose = 1,callbacks=[early_stop],validation_data=(X_test,y_test))
```

```python
predictions = model.predict_classes(X_test)
```

```python
print(classification_report(y_test,predictions))
print('\n')
print(confusion_matrix(y_test,predictions))
```
